```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npr = np.random.rand
```

```python
x, y = npr(100),npr(100)
```

```python
m = 0
c = 0
lr = 0.5
for i in range(1000):
    y_pred = m * x + c
    cost = np.sum((y_pred - y)**2)/(2 * x.shape[0])
    print(cost)
    dm = np.sum((y - y_pred)*(-x))/(x.shape[0])
    dc = np.sum((y - y_pred)*(-1))/(x.shape[0])
    m = m - lr * dm
    c = c - lr * dc

```
